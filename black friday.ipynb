## Problem Statement
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
df_test=pd.read_csv('test.csv')
df_test.head() # it doesnt have the purchase column
```

```python
# append the two data sets
df=df_train.append(df_test)# append method is used cause we want to add the test in the end of the train 
df.head()
```

```python
# basic
df.info()
```

```python
df.describe()
```

```python
# user id is not used so we drop it
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#fixing categorial using map for gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
# handling categorial feature age
df['Age'].unique() # how many unique category do we have
```

```python
#using encoding method beacuse the order of age can help the trian model bethher
df['Age']  =df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})  
```

```python
#the other way is to use label encoding 
from sklearn import preprocessing 
#label_encoder object knows how to understand word labels.
label_encoder=preprocessing.LablEncoder()
#Encode lables in columns 'species'
df['Age']=labe_encoder.fit_transform(df['Age'])
df['Age'].unique()
```

```python
df.head()
```

```python
#fixing categoriacal city category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df=pd.concat([df,df_city],axis=1)
```

```python
df.head()
```

```python
#drop city category feature
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
# missing values
df.isnull().sum() # the ones in the purchase is beacuse the test is also here
```

```python
#for the categorial fetures the best for missing value is to replace the mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
# category 3 replace missing values
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```
