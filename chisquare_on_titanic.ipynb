```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.info()
```

```python
##[sex, embarked, alone, survived, pclass]
df=df[['sex','embarked','alone','survived','pclass']]
df.head()
```

```python
import numpy as np
```

```python
##label encoding for sex feature
df['sex']=np.where(df['sex']=='male',1,0)
##label encoding on embarked
ordinal_label={k:i for i,k in enumerate(df['embarked'].unique(),0)}
df['embarked']=df['embarked'].map(ordinal_label)
```

```python
ordinal_label
```

```python
##label encoding for alone feature
df['alone']=np.where(df['alone']==True,1,0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['sex','embarked','alone','pclass']],df['survived'],
                                               test_size=0.3,random_state=1)
```

```python
X_train.head()
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import chi2
f_pvalues=chi2(X_train,y_train)
```
